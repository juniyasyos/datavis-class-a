## 2. Pengenalan Matplotlib

Matplotlib merupakan fondasi banyak pustaka visualisasi Python. Ia menyediakan kendali penuh terhadap setiap elemen grafis, mulai dari ukuran kanvas hingga warna minor ticks. Memahami struktur Matplotlib membantu Anda menyesuaikan grafik pada level detail.

Beberapa konsep penting:
- `Figure` adalah kanvas besar yang menampung satu atau lebih `Axes`.
- `Axes` adalah area plot tempat data digambar; inilah yang sering dianggap sebagai "plot".
- Dua gaya penulisan yang umum dipakai: pyplot interface (`plt.plot`) dan object-oriented interface (`ax.plot`). Keduanya setara, namun interface berbasis objek lebih fleksibel saat mengelola banyak axes.

````python

# Plot garis sederhana
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("Plot Sine Function")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
````

### Catatan Pembelajaran

- `np.linspace` membuat 100 titik merata dari 0 hingga 10 sehingga kurva terlihat halus.
- Memanggil `plt.figure()` sebelum `plt.plot` memastikan setiap plot berada pada figure baru.
- Jalankan sel dan coba ubah fungsi ke `np.cos` atau ganti warna untuk melihat dampak parameter terhadap gaya visual.

### 2.1 Kustomisasi Dasar Matplotlib

Setiap elemen plot dapat diatur: warna, gaya garis, marker, hingga legenda. Dengan memahami parameter dasar `plt.plot`, Anda dapat membedakan seri data dan menekankan pesan utama grafik. Contoh berikut memperlihatkan beberapa pola kustomisasi yang sering digunakan saat mengajar di kelas.

````python

# Membuat beberapa garis dengan gaya berbeda
y_cos = np.cos(x)

plt.figure()
plt.plot(x, y, label="sin(x)", color="#1f77b4", linewidth=2)
plt.plot(x, y_cos, label="cos(x)", linestyle="--", color="#ff7f0e")
plt.axhline(0, color="black", linewidth=0.8)
plt.title("Perbandingan Sinus dan Cosinus")
plt.xlabel("x")
plt.ylabel("Nilai fungsi")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

### Eksperimen Kustomisasi

- Kombinasi parameter `color`, `linestyle`, dan `marker` membantu membedakan seri data ketika ditampilkan bersamaan.
- `plt.legend()` menampilkan keterangan seri; selalu pastikan label informatif agar audiens cepat memahami makna warna atau marker.
- Cobalah menambahkan argumen `linewidth` atau `markersize` untuk menunjukkan penekanan tertentu pada data.

### 2.2 Mengatur Batas Sumbu, Teks, dan Anotasi

Label yang jelas dan anotasi yang tepat membuat plot lebih komunikatif. Mengatur batas sumbu membantu fokus pada rentang nilai yang relevan, sementara anotasi (`plt.annotate`) menyorot poin penting agar audiens tidak melewatkannya.

````python

plt.figure()
plt.plot(x, y, color="#2ca02c")
plt.title("Mengatur Batas Sumbu dan Anotasi")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)
plt.annotate(
    "Puncak pertama",
    xy=(np.pi / 2, 1),
    xytext=(2, 1.1),
    arrowprops=dict(arrowstyle="->", color="gray")
)
plt.show()
````

### Insight Visual

- `plt.xlim` dan `plt.ylim` mengarahkan perhatian ke rentang nilai yang relevan, mirip seperti melakukan zoom manual.
- `plt.annotate` mampu memberikan konteks terhadap titik penting, misalnya puncak fungsi atau anomali dalam data riil.
- Gunakan parameter `ha` dan `va` agar anotasi tidak menutupi kurva.

## 3. Subplots & Layout

Dalam praktik analisis data, kita sering memerlukan beberapa visual sekaligus untuk perbandingan. Fungsi `plt.subplots` memudahkan pembuatan grid Axes dalam satu Figure. Atur parameter `figsize`, jumlah baris atau kolom, serta `tight_layout` untuk memastikan judul dan label tidak saling tumpang tindih. Bagian ini menekankan bagaimana menata narasi visual secara keseluruhan, bukan sekadar plot tunggal.

````python

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, y, color="#1f77b4")
axes[0, 0].set_title("Line Plot")

axes[0, 1].scatter(x, y, color="#ff7f0e", s=15)
axes[0, 1].set_title("Scatter Plot")

axes[1, 0].bar(["A", "B", "C"], [5, 3, 7], color="#2ca02c")
axes[1, 0].set_title("Bar Plot")

axes[1, 1].hist(y, bins=15, color="#d62728", alpha=0.8)
axes[1, 1].set_title("Histogram")

fig.suptitle("Contoh Subplots dengan Matplotlib", fontsize=16)
plt.tight_layout()
plt.show()
````

### Analisis Layout

- `plt.subplots(2, 2, figsize=(10, 8))` membuat kanvas 2x2; gunakan `figsize` proporsional agar teks tetap terbaca.
- Setiap entry `axes[row, col]` adalah objek `Axes` independen. Anda dapat menerapkan konfigurasi berbeda pada masing-masing subplot.
- Fungsi `fig.tight_layout()` mengurangi overlap antar subplot. Jika label tetap bertumpukan, atur `wspace` atau `hspace` secara manual.

## 4. Studi Kasus: Visualisasi Data Penguins

Dataset *penguins* dari Seaborn berisi ukuran tubuh tiga spesies penguin dari kepulauan Antartika. Studi kasus ini menunjukkan bagaimana memilih jenis plot berdasarkan pertanyaan analisis: membandingkan rata-rata antar spesies dan melihat hubungan antar variabel. Ingatlah untuk selalu membersihkan data (misalnya menghapus nilai hilang) sebelum divisualisasikan agar plot tidak bias.

````python

penguins = sns.load_dataset("penguins").dropna()
penguins.head()
````

### Mengenal Dataset

- Dataset memiliki kolom numerik (`bill_length_mm`, `flipper_length_mm`, `body_mass_g`) dan kategorik (`species`, `sex`, `island`).
- Data asli mengandung nilai hilang; pada contoh ini kita menghapusnya dengan `dropna()` untuk menjaga kebersihan visual.
- Dorong mahasiswa melakukan eksplorasi awal seperti `penguins.describe()` guna memahami skala dan satuan setiap variabel.

### 4.1 Rata-rata Ukuran Paruh per Spesies

Pertanyaan analisis: spesies mana yang memiliki paruh paling panjang dan paling tebal? Kita agregasikan data dengan `groupby` lalu menggambarkannya dalam bar chart. Perhatikan bagaimana label sumbu dan judul yang informatif membantu audiens memahami konteks tanpa harus menatap kode.

````python

summary_bill = (
    penguins
    .groupby("species")[["bill_length_mm", "bill_depth_mm"]]
    .mean()
    .round(2)
    .reset_index()
)
summary_bill
````

````python

plt.figure()
plt.bar(summary_bill["species"], summary_bill["bill_length_mm"], color="#4c72b0")
plt.title("Rata-Rata Panjang Paruh per Spesies Penguin")
plt.xlabel("Spesies")
plt.ylabel("Bill Length (mm)")
plt.show()
````

### Interpretasi Bar Chart

- Bar chart memudahkan perbandingan rata-rata antar kategori; perhatikan bahwa spesies Gentoo memiliki paruh paling panjang.
- Selalu beri label sumbu dengan satuan agar pembaca memahami konteks numerik.
- Pertimbangkan menambahkan anotasi nilai (`plt.bar_label`) pada materi lanjutan untuk memperjelas selisih antar batang.

### 4.2 Scatter Plot: Hubungan Panjang vs Kedalaman Paruh

Scatter plot efektif untuk melihat pola hubungan antar dua variabel numerik. Kita tambahkan warna berdasarkan spesies dan marker berdasarkan jenis kelamin untuk memperkaya cerita visual. Latih siswa untuk membaca pola klaster dan outlier dari plot ini.

````python

colors = {"Adelie": "#1f77b4", "Chinstrap": "#ff7f0e", "Gentoo": "#2ca02c"}

plt.figure()
for species, data in penguins.groupby("species"):
    plt.scatter(
        data["bill_length_mm"],
        data["bill_depth_mm"],
        label=species,
        color=colors[species],
        alpha=0.7
    )

plt.title("Panjang vs Kedalaman Paruh Penguin")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.legend(title="Species")
plt.show()
````

### Membaca Scatter Plot

- Warna (`color`) dan marker (`marker`) digunakan untuk mewakili dua dimensi kategorik sekaligus tanpa membuat plot berlebihan.
- Pola linear menunjukkan hubungan positif antara panjang dan kedalaman paruh; perhatikan bahwa spesies Gentoo terkonsentrasi di area tertentu.
- Diskusikan potensi outlier bersama mahasiswa: apakah outlier tersebut hasil pengukuran atau perlu investigasi lebih lanjut?

## 5. Pengenalan Seaborn

Seaborn dibangun di atas Matplotlib, tetapi menawarkan API deklaratif yang lebih ringkas dan estetika default yang menarik. Integrasi dengan `pandas.DataFrame` memudahkan kita memilih kolom dengan nama variabel, sementara fungsi `relplot`, `catplot`, dan `displot` menyediakan pipeline eksplorasi yang konsisten.

Gunakan Seaborn ketika Anda ingin:
- Membuat plot statistik dengan sedikit kode (misalnya boxplot, violin, regression plot).
- Menangani data kategorik secara otomatis, termasuk pembuatan legenda.
- Menyusun facet (subplot) berdasarkan kategori tanpa harus mengatur axes secara manual.

### 5.1 Memuat Dataset

Seaborn menyediakan sejumlah dataset contoh untuk latihan. Fungsi `sns.load_dataset` mengunduh dataset berbentuk `DataFrame`. Pada kelas, Anda dapat meminta mahasiswa mengecek `tips.info()` atau `tips.describe()` untuk mengenali tipe data sebelum membuat visual.

````python

tips = sns.load_dataset("tips")
tips.head()
````

### Mengenali Variabel `tips`

- Variabel `total_bill` dan `tip` bersifat numerik, cocok untuk analisis hubungan.
- `sex`, `smoker`, `day`, dan `time` adalah kategori yang dapat dimanfaatkan sebagai hue, col, atau style pada plot Seaborn.
- Mintalah mahasiswa membuat hipotesis: misalnya, apakah tamu malam hari memberi tip lebih besar?

### 5.2 Axis-Level Plot

Axis-level function (misalnya `sns.scatterplot`, `sns.histplot`) mengembalikan objek `Axes`. Ini cocok untuk fokus pada satu visual dan memberi kesempatan untuk memadukan kembali dengan API Matplotlib jika ingin penyesuaian lanjutan, seperti menambahkan garis rata-rata atau anotasi khusus.

````python

fig, ax = plt.subplots()
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time",
    style="sex",
    palette="Set2",
    ax=ax
)
ax.set_title("Hubungan Total Bill dan Tip")
plt.show()
````

### Membaca Hasil Axis-Level Plot

- `sns.scatterplot` secara otomatis membuat legenda berdasarkan parameter `hue` dan `style`, menghemat kode boilerplate.
- Parameter `ax` memungkinkan integrasi dengan Matplotlib; Anda bisa menambahkan garis referensi menggunakan `ax.axhline`.
- Eksperimen: ubah `style` menjadi `smoker` untuk melihat apakah status perokok memengaruhi pola tip.

````python

fig, ax = plt.subplots()
sns.histplot(
    data=tips,
    x="total_bill",
    kde=True,
    bins=20,
    color="#4c72b0",
    ax=ax
)
ax.set_title("Distribusi Total Bill")
plt.show()
````

### Distribusi dan KDE

- Histogram menunjukkan frekuensi total bill, sementara `kde=True` menambahkan kurva densitas yang mulus.
- Gunakan argumen `bins` untuk mengontrol resolusi histogram; terlalu banyak bin dapat membuat pola sulit dibaca.
- Ajak mahasiswa mencoba `multiple='stack'` atau `hue='day'` untuk membandingkan distribusi per hari.

### 5.3 Figure-Level Plot

Figure-level interface (`sns.catplot`, `sns.relplot`, `sns.displot`, dan lain-lain) otomatis membuat `Figure` baru dan mengelola grid subplot berdasarkan parameter `row`, `col`, atau `hue`. Gunakan pendekatan ini ketika Anda ingin membandingkan beberapa subkelompok sekaligus tanpa menulis banyak kode boilerplate.

````python

sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    kind="box",
    height=5,
    aspect=1.2,
    palette="pastel"
).set(title="Sebaran Total Bill per Hari")
plt.show()
````

### Figure-Level di Data Kategorik

- `sns.catplot` otomatis membuat figure baru dan menata legenda. Parameter `kind` dapat diubah ke `violin`, `strip`, atau `point` sesuai kebutuhan analisis.
- Karena catplot bersifat figure-level, gunakan method `set_axis_labels` atau `set_titles` dari objek yang dikembalikan untuk penyesuaian tambahan.
- Tanyakan pada mahasiswa: apa insight yang didapat mengenai perbedaan total bill antara jenis kelamin?

````python

sns.relplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    col="sex",
    kind="scatter",
    height=5,
)
plt.show()
````

### Facet untuk Perbandingan

- `col='sex'` menghasilkan dua kolom subplot sehingga kita bisa membandingkan pola hubungan antar jenis kelamin secara berdampingan.
- Parameter `height` dan `aspect` mengendalikan ukuran masing-masing facet; atur agar teks tidak terlalu padat.
- Gunakan `facet.add_legend()` atau `facet.set_axis_labels` jika ingin memperkaya tampilan sebelum ekspor ke laporan.

### 5.4 Kustomisasi Tema dan Palet

Selain fungsi plotting, Seaborn memiliki utilitas untuk mengelola tampilan. Tema (`set_theme`, `set_style`) mengatur elemen umum seperti latar belakang dan grid, sementara `color_palette` membantu menjaga konsistensi warna di seluruh materi ajar. Manfaatkan konteks (`with sns.axes_style(...)`) untuk sementara waktu mengubah gaya tanpa memengaruhi plot selanjutnya.

````python

with sns.axes_style("darkgrid"):
    fig, ax = plt.subplots()
    sns.kdeplot(
        data=penguins,
        x="bill_length_mm",
        hue="species",
        fill=True,
        common_norm=False,
        alpha=0.6,
        linewidth=1,
        ax=ax
    )
    ax.set_title("Distribusi Panjang Paruh per Spesies")
plt.show()
````

### Mengelola Tema dan Palet

- Blok konteks `with sns.axes_style("darkgrid")` membantu mencoba gaya alternatif tanpa mengubah konfigurasi global.
- `sns.color_palette("flare")` menghasilkan kombinasi warna yang konsisten, penting saat grafik dibagikan ke audiens luas.
- Eksperimen lanjutan: gunakan `sns.set_palette` di awal notebook untuk menerapkan palet tertentu di seluruh plot.

## 6. Matplotlib vs Seaborn: Kapan Menggunakan yang Mana?

- **Matplotlib**: pilih saat Anda membutuhkan kontrol granular, membangun visual custom, atau mempersiapkan grafik untuk publikasi ilmiah dengan spesifikasi ketat.
- **Seaborn**: pilih untuk eksplorasi cepat, analisis statistik standar, serta ketika Anda butuh facet berdasarkan kategori dengan kode singkat.
- Kombinasikan keduanya: buat plot dengan Seaborn untuk memanfaatkan default yang indah, lalu gunakan method Matplotlib (`ax.set_title`, `ax.axhline`, dan seterusnya) untuk penyesuaian akhir.
